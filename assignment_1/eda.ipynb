**Evaluating Most Common Words and Token Frequency**

```python
from collections import Counter
import re

# Minimal function to split into tokens
def simple_tokenizer(text):
    text = text.lower()
    tokens = re.findall(r'\w+', text)
    return tokens

all_text = " ".join(train_df['conversation'].astype(str).tolist()).lower()
tokens = simple_tokenizer(all_text)
most_common = Counter(tokens).most_common(20)
print("Top 20 most common tokens across all data:")
print(most_common)
```

```python
for s in train_df['customer_sentiment'].unique():
    subset = train_df[train_df['customer_sentiment'] == s]
    tokens_sub = simple_tokenizer(" ".join(subset['conversation'].astype(str).tolist()))
    mc = Counter(tokens_sub).most_common(10)
    print(f"Sentiment {s} | Top Tokens: {mc}")
```

**Evaluating Correlations**

```python
#Encoding Sentiment into Numerical Value
sentiment_map = {"negative": 0, "neutral": 1, "positive": 2}
train_df["sentiment_num"] = train_df["customer_sentiment"].map(sentiment_map)
```

```python
#Chi-Square Test for Agent Experience Level and Sentiment
import scipy.stats as stats

cont_table_ael_s = pd.crosstab(train_df["agent_experience_level"], #Contingency Table
                         train_df["customer_sentiment"])
print(cont_table_ael_s)
```

```python
chi2, p, dof, expected = stats.chi2_contingency(cont_table_ael_s)
print(f"Chi-square statistic: {chi2:.4f}")
print(f"p-value: {p:.4g}")
```

Since p-value > 0.05, we can not say that there is a statiscally significant relationship between agent experience level and customer sentiment.

```python
#Chi-Square Test for Issue Complexity and Sentiment
cont_table_ico_s = pd.crosstab(train_df["issue_complexity"], #Contingency Table
                         train_df["customer_sentiment"])
print(cont_table_ico_s)
```

```python
chi2, p, dof, expected = stats.chi2_contingency(cont_table_ico_s)
print(f"Chi-square statistic: {chi2:.4f}")
print(f"p-value: {p:.4g}")
```

Since p-value > 0.05, we can not say that there is a statiscally significant relationship between issue complexity and customer sentiment.

```python
#Chi-Square Test for Issue Category and Sentiment
cont_table_ica_s = pd.crosstab(train_df["issue_category"], #Contingency Table
                         train_df["customer_sentiment"])
print(cont_table_ica_s)
```

```python
chi2, p, dof, expected = stats.chi2_contingency(cont_table_ica_s)
print(f"Chi-square statistic: {chi2:.4f}")
print(f"p-value: {p:.4g}")
```

Since p-value much smaller than 0.05, we can say that there is a strong and highly significant relationship between issue category and customer sentiment.
